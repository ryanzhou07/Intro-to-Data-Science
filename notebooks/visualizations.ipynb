```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
df = pd.read_csv('Data/monthy_player_count.csv')
df2 = pd.read_csv('Data/Reviews.csv')
```

```python
fig1, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

sns.histplot(df['Avg_Players'], bins=15, ax=ax1, color='steelblue')
ax1.set_title('Average Player Counts (Linear)')
ax1.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.setp(ax1.get_xticklabels(), rotation=45)

sns.histplot(df['Avg_Players'], bins=15, ax=ax2, color='indianred', log_scale=True)
ax2.set_title('Average Player Counts (Log Scale)')

fig1.tight_layout()
fig1.savefig('Graphs/players_distribution.pdf', bbox_inches='tight') 
plt.show()

fig2, (ax3, ax4) = plt.subplots(1, 2, figsize=(10, 5))

sns.histplot(df2['total_reviews'], bins=15, ax=ax3, color='darkslateblue')
ax3.set_title('Total Reviews (Linear)')
ax3.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.setp(ax3.get_xticklabels(), rotation=45)

sns.histplot(df2['total_reviews'], bins=15, ax=ax4, color='seagreen', log_scale=True)
ax4.set_title('Total Reviews (Log Scale)')

fig2.tight_layout()
fig2.savefig('Graphs/reviews_distribution.pdf', bbox_inches='tight')
plt.show()
```
