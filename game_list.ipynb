```python
import requests
import pandas as pd
import time
```

```python
# 1. Setup your credentials and parameters
MY_API_KEY = "API HERE"

def get_complete_steam_catalog(api_key):
    base_url = "https://api.steampowered.com/IStoreService/GetAppList/v1/"
    all_apps = []
    last_appid = 0  # We start at the beginning
    
    while True:
        params = {
            'key': api_key,
            'max_results': 50000,
            'last_appid': last_appid, # This tells Steam where to start the next page
            'include_games': True
        }
        
        print(f"Fetching page starting after AppID: {last_appid}...")
        response = requests.get(base_url, params=params).json()
        
        # Check if we got data back
        apps = response.get('response', {}).get('apps', [])
        if not apps:
            break
            
        all_apps.extend(apps)
        
        # Update last_appid to the last one we just received
        last_appid = apps[-1]['appid']
        
        # Optional: Add a tiny sleep to avoid rate limits (200 requests / 5 mins)
        time.sleep(1) 

    df = pd.DataFrame(all_apps)
    df.to_csv('complete_steam_lookup_2026.csv', index=False)
    print(f"Done! Saved {len(df)} games.")

get_complete_steam_catalog(MY_API_KEY)
```
